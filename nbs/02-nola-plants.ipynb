```python
import imageio.v3 as iio
from glitchpy import glitchpy as gpy
from glitchpy import view
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
from skimage import color
%load_ext autoreload
%autoreload 2
```

```python
imgs_path = Path(r'C:\Users\cgusb\Pictures\photos\NOLA')
imgs = []
for path in imgs_path.glob('*.jpg'):
    imgs.append(iio.imread(path))
```

```python
imgs[0].shape
```

```python
img = imgs[np.random.randint(0, len(imgs))]
fig, ax = plt.subplots()
ax.imshow(img)
ax.axis('off')
plt.show()
```

```python
fig, ax = view.image(img, dpi=150)
plt.show()
fig, axes = view.channels(img, dpi=150)
plt.show()
```

```python
img_hsv = color.rgb2hsv(img)
fig, ax = view.image(img_hsv, dpi=150)
plt.show()
fig, ax = view.channels(img_hsv, dpi=150, chan_range=[0, 1])
plt.show()
```

```python
fig, ax = view.histogram(img, channel_labels='HSV')
plt.show()
```
